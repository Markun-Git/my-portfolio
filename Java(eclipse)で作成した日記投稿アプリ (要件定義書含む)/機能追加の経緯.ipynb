# 📝 Amigos Diary_neo シリーズ変更履歴

## 🔄 Amigos Diary_neo の変更点
- ログイン時のユーザー名と、ログイン後に表示されるユーザー名を区別
- 会員登録機能を追加

## 💬 Amigos Diary_neo2 の変更点
- 投稿された日記に**コメント機能**を追加（※ログインユーザー限定）
  - 自分のコメントは後から**編集・削除可能**
  - **日記が削除されるとコメントも削除**（ただしDB上には残る）

## 👍 Amigos Diary_neo3 の変更点
- 日記に「**いいね👍**」ボタンを追加
  - 押した人数がカウントされる
  - 同じユーザーが2回押すことはできない
  - 再度押すとカウントが減り**いいねが解除**される

## 👋 Amigos Diary_neo4 の変更点
- **会員退会機能**を追加
  - 会員情報、投稿日記、コメントも削除（※DB上には残る）
  - `FLAG`カラムが `1 → 0` に更新されるだけ
  - いいね数はそのまま残る

## 🔍 Amigos Diary_neo5 の変更点
- **日記検索機能**を追加
  - 検索窓を日記一覧画面に追加
  - 含まれる文字を検索できる
  - 空白で検索するとメッセージ表示

## ⭐ Amigos Diary_neo6 の変更点
- **お気に入りユーザー（フォロー）機能**を追加
  - フォローしたユーザーの日記が「お気に入り」から確認可能
- **会員登録情報編集機能**を追加
  - アイコンとひとことメッセージの編集・登録が可能に

## 🛠 Amigos Diary_neo7 の変更点
- 新規テーブルの追加は無し
- 会員情報変更画面で**ログインネームやパスワードの変更**が可能に
- 日記一覧で**自分の投稿を紫色で強調表示**

## 📄 Amigos Diary_neo8 の変更点
- 新規テーブルの追加は無し
- **ページ分割機能**を追加（10件以上投稿がある場合）
- アイコン画像を会員登録時のものと一致させるよう修正
  - 新しい日記投稿画面ではアイコン選択ができないように変更

## 🎥 Amigos Diary_neo9 の変更点
- 新規テーブルの追加は無し
- テーブルのCSSデザインを元に戻す
- **YouTube埋め込み機能**を追加
  - 投稿本文に埋め込みコードを貼り付けることで動画投稿が可能に
  - 画像との**同時投稿も可**
  - 投稿・編集時に貼り方の案内を表示

## 🎨 Amigos Diary_neo10 の変更点
- 新規テーブルの追加は無し
- **検索結果画面の挙動を改善**
  - 会員アイコンが正しく表示されるよう修正
- CSS追加で背景を反映
  - 背景は薄く表示、文字が見やすくなるよう調整

## 🎆 Amigos Diary_neo11 の変更点
- 新規テーブル・新機能の追加は無し
- CSS・JavaScriptで**演出とデザインを大幅改善**
  - ログイン成功・ログアウト時に花火や花吹雪演出
  - テーブルやフォントの視認性向上
  - 画面表示時に**スムーズなフェード表示**を追加
